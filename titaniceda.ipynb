```python
# Exploratory Data Analysis (EDA) on Sales Dataset

#This notebook performs a comprehensive exploratory data analysis on the Sales dataset,
#including data overview, univariate and bivariate analysis, correlation study, outlier detection,
#missing value handling, and summary of insights.
```

```python
# Step 1: Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Settings for plots
sns.set(style="whitegrid")
%matplotlib inline
```

```python
# Step 2: Load Dataset

df = pd.read_csv('train.csv')  # update filename/path if needed
df.head()
```

```python
# Step 3: Data Overview

# Basic info
print("Dataset Information:")
df.info()

# Statistical summary
print("\nStatistical Summary:")
print(df.describe())

# Checking missing values
print("\nMissing Values Count:")
print(df.isnull().sum())

# Value counts for key categorical columns
print("\nValue Counts for Ship Mode:")
print(df['Ship Mode'].value_counts())

print("\nValue Counts for Segment:")
print(df['Segment'].value_counts())

print("\nValue Counts for Category:")
print(df['Category'].value_counts())
```

```python
# Step 4: Univariate Analysis
# Sales Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Sales'].dropna(), bins=50, kde=True, color='skyblue')
plt.title("Sales Distribution")
plt.xlabel("Sales Amount")
plt.ylabel("Frequency")
plt.show()
# Ship Mode Distribution
plt.figure(figsize=(6,4))
sns.countplot(x='Ship Mode', data=df, palette='Set2')
plt.title("Ship Mode Distribution")
plt.xlabel("Ship Mode")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
# Customer Segment Distribution
plt.figure(figsize=(6,4))
sns.countplot(x='Segment', data=df, palette='Set1')
plt.title("Customer Segment Distribution")
plt.xlabel("Segment")
plt.ylabel("Count")
plt.show()
```

```python
# Step 5: Bivariate Analysis
# Sales by Ship Mode
plt.figure(figsize=(8,5))
sns.boxplot(x='Ship Mode', y='Sales', data=df, palette='Set3')
plt.title("Sales by Ship Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Sales")
plt.show()
# Sales by Segment
plt.figure(figsize=(8,5))
sns.boxplot(x='Segment', y='Sales', data=df, palette='Set2')
plt.title("Sales by Customer Segment")
plt.xlabel("Segment")
plt.ylabel("Sales")
plt.show()
# Sales by Category
plt.figure(figsize=(8,5))
sns.boxplot(x='Category', y='Sales', data=df, palette='Set1')
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
# Sales by Region and Segment (Grouped Bar Plot)
plt.figure(figsize=(10,6))
sns.barplot(x='Region', y='Sales', hue='Segment', data=df, ci=None, palette='pastel')
plt.title("Average Sales by Region and Segment")
plt.xlabel("Region")
plt.ylabel("Average Sales")
plt.show()
```

```python
# Step 6: Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Step 7: Outlier Detection (Boxplots)
plt.figure(figsize=(8,4))
sns.boxplot(x=df['Sales'])
plt.title("Boxplot for Sales")
plt.show()
```

```python
# Step 8: Handling Missing Values
# Check missing again for clarity
print(df.isnull().sum())

# Example: Fill missing Sales with median (if any)
df['Sales'].fillna(df['Sales'].median(), inplace=True)

# Example: Fill missing Ship Mode with mode (if any)
df['Ship Mode'].fillna(df['Ship Mode'].mode()[0], inplace=True)

# Verify no missing now
print(df.isnull().sum())
```

```python
# Step 9: Feature Analysis Examples
# Total Sales by Sub-Category
plt.figure(figsize=(12,5))
sns.barplot(x='Sub-Category', y='Sales', data=df, estimator=sum, ci=None, palette='muted')
plt.title("Total Sales by Sub-Category")
plt.xlabel("Sub-Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
# Average Sales in Top 10 States
top_states = df['State'].value_counts().nlargest(10).index
plt.figure(figsize=(10,5))
sns.barplot(x='State', y='Sales', data=df[df['State'].isin(top_states)], estimator=np.mean, ci=None, palette='Set2')
plt.title("Average Sales in Top 10 States")
plt.xlabel("State")
plt.ylabel("Average Sales")
plt.xticks(rotation=45)
plt.show()
```
